```python
from io_utils import read_fbin, read_ibin
file_path = "/users/dellatod/yandex/t2l1b/base.1M.fbin"
query_path = "/users/dellatod/yandex/t2l1b/query.public.100K.fbin"
ground_truth_path = "/users/dellatod/yandex/t2l1b/groundtruth.public.100K.ibin"
data_vector = read_fbin(file_path)
```
